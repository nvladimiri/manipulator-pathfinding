### Libraries 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
from matplotlib.animation import FuncAnimation

import import_ipynb

from Astar import Map, astar, manhattan_distance, SearchTreePQD, make_path
from CppAlgorithms import get_state_cells, rotate_points_partial, get_motion_cells, generate_animation
```

### Input

```python
n = 8  # number of divisions for angles
echelon_points = [(0, 0), (0, 50), (50, 50), (100, 100)]  # coordinates of the vertices of the segments
obstacles = [(x, y) for x in range(-40, -9) for y in range(50, 81)] + [(110, 110), (80, 50)]  # coordinates of obstacles
start = [0, 0, 0]  # start coordinates (indexed by division)
target = [3, 0, 0]  # target coordinates (indexed by division)

if not(len(echelon_points) == len(start) + 1 == len(target) + 1):
    raise ValueError("The number of segments should be one less than the number of vertices of the segments")
```

### Presentation

```python
path_length = None
graph = Map(n, echelon_points, obstacles)
algorithm = astar(n, graph, start, target, manhattan_distance, SearchTreePQD)
result, last_node, steps = algorithm[0],  algorithm[1], algorithm[2]

if result:
    path, path_length = make_path(last_node)
    print(f"result = {result},\npath_length = {path_length},\nsteps = {steps}")
    generate_animation(path, echelon_points, obstacles, n)
else:
    print(f"Path doesn't exists")
    
```
